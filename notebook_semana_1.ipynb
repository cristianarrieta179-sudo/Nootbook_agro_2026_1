```python
# ENTRAR entra ancho alto de la materia
ancho = input("ingresa el ancho de la ventana en metros ")
ancho = float(ancho)
alto = input("ingresa el alto de la ventana en metros")
alto = float(alto)


# PROCESO calcular longitud de la madera (lm) y el area del vidreo
lm= 2 * (alto * ancho) * 1.26
ar= 2 * (alto * ancho)

# SALIDA sale la longitud de la madera en (lm) y el area del vidreo en metros cuadrados
print("longitud de la madera" ,lm , "sale la longitud")
print("el area del vidreo",ar, "en metros cuadrados")

```
