### Ejemplos Sencillos de Python en Agroindustria

Aquí tienes tres aplicaciones prácticas con código básico:

#### 1. Cálculo del Índice de Madurez

Este pequeño programa ayuda a determinar si una fruta está lista para proceso basándose en la relación entre sólidos solubles ($^\circ Brix$) y acidez.

```
# Cálculo de relación de madurez (SST/Acidez)
brix = 12.5
acidez = 0.8

indice_madurez = brix / acidez

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Óptimo para cosecha/proceso.")
else:
    print("Estado: Aún le falta maduración.")
```

#### 2. Monitoreo de Temperatura en Almacenamiento

Imagina un script que recibe datos de un sensor y emite una alerta si la cadena de frío se rompe.

```
# Simulación de sensor de temperatura en un silo
temperatura_actual = 5.2  # grados Celsius
limite_maximo = 4.0

if temperatura_actual > limite_maximo:
    diferencia = temperatura_actual - limite_maximo
    print(f"¡ALERTA! La temperatura excedió el límite por {diferencia}°C.")
else:
    print("Temperatura estable. Producto seguro.")
```

#### 3. Control de Inventario de Materia Prima

Un ejemplo de cómo calcular el rendimiento de una línea de producción tras el descarte.

```
# Cálculo de rendimiento de proceso
peso_inicial = 500  # kg de materia prima bruta
descarte = 45       # kg de cáscaras, semillas o merma

peso_final = peso_inicial - descarte
rendimiento = (peso_final / peso_inicial) * 100

print(f"Peso neto obtenido: {peso_final} kg")
print(f"El rendimiento total del proceso fue del: {rendimiento}%")
```
